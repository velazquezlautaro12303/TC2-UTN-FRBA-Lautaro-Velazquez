```python
%matplotlib ipympl
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt

# Definir el orden del filtro y la frecuencia de corte
orden = 1
frecuencia_corte = 1.0  # En radianes/segundo

# Dise√±ar el filtro Bessel
b, a = signal.bessel(orden, frecuencia_corte, analog=True)

# Mostrar los coeficientes del numerador (b) y del denominador (a)
print("Coeficientes del numerador (b):", b)
print("Coeficientes del denominador (a):", a)

# Crear una funci√≥n de transferencia
sistema = signal.TransferFunction(b, a)

# Mostrar la funci√≥n de transferencia
print("\nFunci√≥n de transferencia:")
print(sistema)

# Graficar la respuesta en frecuencia
w, mag, phase = signal.bode(sistema)

plt.figure()
plt.subplot(2, 1, 1)
plt.semilogx(w, mag)
plt.title('Respuesta en Frecuencia del Filtro Bessel de Primer Orden')
plt.ylabel('Magnitud (dB)')
plt.grid()

plt.subplot(2, 1, 2)
plt.semilogx(w, phase)
plt.xlabel('Frecuencia (rad/s)')
plt.ylabel('Fase (grados)')
plt.grid()

plt.show()
```
